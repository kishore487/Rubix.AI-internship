```python
# List of columns to check for outliers
columns_to_plot = ['LotFrontage', 'LotArea', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2',
                   'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'GrLivArea']

# Loop through each column and remove outliers
for column in columns_to_plot:
    Q1 = df[column].quantile(0.25)  # 25th percentile
    Q3 = df[column].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1                   # Interquartile Range
    
    # Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter the DataFrame to exclude outliers
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import pandas as pd

# One-hot encode categorical variables
df_encoded = pd.get_dummies(df)

# Splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(df_encoded.drop(['SalePrice'], axis=1), df_encoded['SalePrice'], test_size=0.2, random_state=42)

models = {
    'RF': RandomForestRegressor(),
    'XGB': XGBRegressor()
}

for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    print(f"{name} - MAE: {mae}, MSE: {mse}, RMSE: {rmse}, RÂ²: {r2}")
```

Both models have similar performance metrics, with the Random Forest Regressor slightly outperforming the XGBoost Regressor in terms of MAE, MSE, RMSE, and RÂ² score.

The high RÂ² scores (above 0.85) indicate that both models explain a significant portion of the variance in the target variable, SalePrice.

The relatively low MAE and RMSE values suggest that the models' predictions are close to the actual values.
